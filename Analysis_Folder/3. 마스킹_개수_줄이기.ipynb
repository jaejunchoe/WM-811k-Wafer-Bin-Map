```python
# 'Train의 failureType_str에 따라 그룹화하고 각 그룹의 크기를 계산합니다
label_counts = df_tr['failureType_str'].value_counts()

# 결과 출력
print(label_counts)
```

```python
# 'Test의 failureType_str에 따라 그룹화하고 각 그룹의 크기를 계산합니다
label_counts = df_te['failureType_str'].value_counts()

# 결과 출력
print(label_counts)
```

```python
# df_tr에서 'none' 타입 행을 추출 후 3000개로 제한
df_tr_none = df_tr[df_tr['failureType_str'] == "[['none']]"]
df_tr_none_limited = df_tr_none.sample(n=3000, random_state=42)  # 무작위로 3000개 샘플링

# df_te에서 'none' 타입 행을 추출 후 3000개로 제한
df_te_none = df_te[df_te['failureType_str'] == "[['none']]"]
df_te_none_limited = df_te_none.sample(n=3000, random_state=42)  # 무작위로 3000개 샘플링

# 나머지 데이터 추출
df_tr_rest = df_tr[df_tr['failureType_str'] != "[['none']]"]
df_te_rest = df_te[df_te['failureType_str'] != "[['none']]"]

# 제한된 'none' 데이터와 나머지 데이터를 병합
df_tr_final = pd.concat([df_tr_none_limited, df_tr_rest])
df_te_final = pd.concat([df_te_none_limited, df_te_rest])

# 인덱스 재설정
df_tr_final.reset_index(drop=True, inplace=True)
df_te_final.reset_index(drop=True, inplace=True)

# 결과 확인
print(df_tr_final['failureType_str'].value_counts())
print(df_te_final['failureType_str'].value_counts())
```

```python
# df_tr_final 데이터프레임을 pickle 파일로 저장
df_tr_final.to_pickle("/content/drvie/MyDrive/반도체_캐글/LSWMD/failure_df_Tr.pkl")

# df_te_final 데이터프레임을 pickle 파일로 저장
df_te_final.to_pickle("/content/drvie/MyDrive/반도체_캐글/LSWMD/failure_df_Te.pkl")
```

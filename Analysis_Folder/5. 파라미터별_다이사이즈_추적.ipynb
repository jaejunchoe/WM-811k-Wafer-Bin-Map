```python
import pandas as pd
import numpy as np

# df_cleaned는 이미 있는 데이터프레임이라고 가정합니다.

# waferMap_parameter가 (26,26)인 행만 필터링
filtered_df = df[df['waferMap_parameter'] == (26, 26)]

# dieSize와 failureType에 따라 그룹화
grouped = filtered_df.groupby(['dieSize', 'failureType'])

# 각 dieSize와 failureType 그룹 내에서 trainTestLabel의 빈도 계산
for (die_size, failure_type), group in grouped:
    trainTestLabel_counts = group['trainTestLabel'].value_counts()

    print(f"\ndieSize가 {die_size}, failureType이 {failure_type}인 경우:")
    print("trainTestLabel의 값과 개수:")
    print(trainTestLabel_counts)
```
